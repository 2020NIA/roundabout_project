# 사상자 수

```python
data=pd.read_csv('train_boxcox.csv',encoding='cp949')
test = pd.read_csv('test_boxcox.csv',encoding='cp949')
test.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
model = data.drop(['위도','경도'],axis=1)
```

```python
circle_idx = model[model.교차로명.str.contains('회전교차로')].index
before = model.drop(index=circle_idx,axis=0)['person']
```

```python
model = model.set_index('교차로명')

train = pd.get_dummies(model)

train.drop(['speed','peek_speed','car','bus','truck','accident'],axis=1,inplace=True)
```

```python
train.head()
```

```python
test_name = test['교차로명']

test=test.set_index('교차로명')

test = pd.get_dummies(test)

test.drop(['speed','peek_speed','car','bus','truck','accident'],axis=1,inplace=True)

train_x = train.drop('person',axis=1)
train_y = train['person']


test_x = test.drop('person',axis=1)
```

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
rf = RandomForestRegressor()
gra = GradientBoostingRegressor()
rf.fit(train_x,train_y)
gra.fit(train_x,train_y)
rf_pred = rf.predict(test_x)
gra_pred = gra.predict(test_x)
```

```python
person_table=pd.DataFrame({'교차로명':test_name,'before':before})
person_table['rf_pred'] = rf_pred
person_table['gra_pred'] = gra_pred
```

```python
person_table['avg'] = (person_table['rf_pred'] + person_table['gra_pred'])/2
```

```python
person_table['avg'] = np.ceil(person_table['avg'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_columns',40)
```

```python
data= pd.read_csv('원형교차로_사고.csv')
data.head()
```

```python
data.columns
```

```python
data['준공연도'].unique()
```

```python
data['스쿨존 내\n위치여부'].fillna('X',inplace=True)
```

```python
def o2o(x):
    if x != 'X':
        return 'O'
    else:
        return x
data['스쿨존 내\n위치여부'] = data['스쿨존 내\n위치여부'].apply(lambda x: o2o(x))
```

```python
data['교차로 내 신호 여부'].fillna('X',inplace=True)
data['교차로 내 신호 여부'] = data['교차로 내 신호 여부'].apply(lambda x: o2o(x))
```

```python
data['접근로 진입부\n양보선 여부'].fillna('X',inplace=True)
data['접근로 진입부\n양보선 여부'] = data['접근로 진입부\n양보선 여부'].apply(lambda x: o2o(x))
```

```python
# 159 로터리
data['시도'].unique()
```

```python
data.drop(['시군구','도로명','기타주소','도로등급','시행청','계','신규추가','누락추가','비고','위도','경도'],axis=1,inplace=True)
```

```python
data.columns = ['교차로명','시도', '준공연도', '교차로 유형', '회전교차로 유형', '로터리 유형', '스쿨존 내 위치여부','교차로 내 신호 여부', '접근로 진입부 양보선 여부', '사고 건수']
```

```python
로터리 = data[data['교차로명'].str.contains('로터리')]
```

```python
로터리.head()
```

```python
lottery = 로터리.reset_index(drop=True)
```

```python
for i in range(lottery.shape[0]):
    if type(lottery['회전교차로 유형'][i]) != float:
        x=lottery['회전교차로 유형'][i]
        lottery['로터리 유형'][i] = x
    else:
        continue
lottery
```

```python
lottery.drop(['회전교차로 유형'],axis=1,inplace=True)
```

```python
lottery.columns
```

```python
lottery['로터리 유형']=lottery['로터리 유형'].str[:3]
lottery['로터리 유형'].unique()
```

```python
lottery.drop('준공연도',axis=1,inplace=True)
```

```python
lottery.to_csv('로터리 사고 건수.csv',index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from statsmodels.formula.api import ols
```

```python
x_test = pd.read_csv("/Users/yangdongjae/Desktop/2020/대외활동/2020년 공공 빅데이터 청년 인턴십/실무형 프로젝트/roundabout_project/data/X_test.csv")
x_train = pd.read_csv("/Users/yangdongjae/Desktop/2020/대외활동/2020년 공공 빅데이터 청년 인턴십/실무형 프로젝트/roundabout_project/data/X_test.csv")
```

```python
x_train.set_index('교차로명',inplace = True)
x_test.set_index('교차로명',inplace = True)
```

```python
y_train = x_train['사고건수']
x_train.drop(columns = ['사고건수'], inplace = True)
```

```python
y_test = x_test['사고건수']
x_test.drop(columns = ['사고건수'], inplace = True)
```

```python
min_max = MinMaxScaler()
```

```python
x_train.columns
```

```python
min_max_tr = min_max.fit_transform(x_train.drop(columns = ['좌회전8', '좌회전13', '좌회전18']))
```

```python
x = pd.DataFrame(min_max_tr)
```

```python
x.columns = ['통행량8', '통행량13', '통행량18']
```

```python
x.columns
```

```python
x.reset_index(drop = True, inplace = True)
x_train.reset_index(drop = True, inplace = True)
```

```python
x_train
```

```python
train_x = pd.concat([x, x_train.iloc[:,3:6]], axis = 1)
```

```python
train_x
```

```python
train = pd.concat([train_x,y_train.reset_index(drop = True)],axis = 1)
```

```python
train.head()
```

```python
res = ols('사고건수 ~ 통행량8 + 통행량13 + 통행량18 + 좌회전8 + 좌회전13 + 좌회전18', data = train).fit()
```

```python
res.summary()
```
